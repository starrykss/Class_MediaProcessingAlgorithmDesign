# Histogram Stretching

#### OpenCV에서 함수로 제공하지 않아 직접 구현해야 함.

$$
dist(x, y) = \frac{src(x,y)-G_{min}}{G_{max} - G_{min}} \times 255
$$

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
src = cv2.imread('./data/hawkes.bmp', cv2.IMREAD_GRAYSCALE)
img = cv2.imread('./data/hawkes.bmp')
plt.figure(figsize=(7, 7))
plt.axis('off')
plt.imshow(img)
plt.title('Original Image')
plt.show()
```

```python
plt.hist(src.flatten(), range=[0,256], bins=256)
plt.title('Historgram')
plt.show()
```

```python
# TODO: Histogram Stretching

min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(src)
print('min = ', min_val)
print('max = ', max_val)


dst = ((src-min_val)*255) / (max_val-min_val) # Histogram Stretching
dst = cv2.convertScaleAbs(dst)

plt.hist(dst.flatten(), range=[0,256], bins=256)
plt.title('Histogram')
plt.show()

cv2.imshow('original', src)
cv2.imshow('stretching', dst)
cv2.waitKey()
cv2.destroyAllWindows()
```
