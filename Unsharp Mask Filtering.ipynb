# Unsharp Mask Filtering


### : 블러링이 적용된 날카롭지 않은 영상을 이용하여 날카로운 영상으로 만드는 필터링

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#1
src = cv2.imread('./data/lena.jpg', cv2.IMREAD_GRAYSCALE)
blur = cv2.GaussianBlur(src, ksize=(7, 7), sigmaX=0.0)

#2
fx = src
f_x = blur
gx = fx - f_x
hx = fx + gx

# 히스토그램 표시
hist1 = (cv2.calcHist(images=[fx], channels=[0], mask=None, histSize=[256], ranges=[0, 256])).flatten()
hist2 = (cv2.calcHist(images=[f_x], channels=[0], mask=None, histSize=[256], ranges=[0, 256])).flatten()
hist3 = (cv2.calcHist(images=[gx], channels=[0], mask=None, histSize=[256], ranges=[0, 256])).flatten()
hist4 = (cv2.calcHist(images=[hx], channels=[0], mask=None, histSize=[256], ranges=[0, 256])).flatten()


plt.plot(hist1, color='b')
plt.title('fx')
plt.show()

plt.plot(hist2, color='b')
plt.title('f_x')
plt.show()

plt.plot(hist3, color='b')
plt.title('gx')
plt.ylim(0, 400)
plt.show()

plt.plot(hist4, color='b')
plt.title('hx')
plt.show()

fx = cv2.cvtColor(fx, cv2.COLOR_BGR2RGB)
f_x = cv2.cvtColor(f_x, cv2.COLOR_BGR2RGB)
gx = cv2.cvtColor(gx, cv2.COLOR_BGR2RGB)
hx = cv2.cvtColor(hx, cv2.COLOR_BGR2RGB)

fig, ax = plt.subplots(2, 2, figsize = (7, 7), sharey = True)   # 2 by 2, inch, 각 이미지 좌표 Share 여부

ax[0][0].axis('off')
ax[0][0].imshow(fx, aspect = 'auto')

ax[0][1].axis('off')
ax[0][1].imshow(f_x, aspect = 'auto')

ax[1][0].axis('off')
ax[1][0].imshow(gx, aspect = 'auto')

ax[1][1].axis('off')
ax[1][1].imshow(hx, aspect = 'auto')

plt.subplots_adjust(left = 0, bottom = 0, right = 1, top = 1, wspace = 0.05, hspace = 0.05)
plt.show()

'''
cv2.imshow('fx', fx)
cv2.imshow('f_x', f_x)
cv2.imshow('gx', gx)
cv2.imshow('hx', hx)
'''
cv2.waitKey()
cv2.destroyAllWindows()
```
